```python
# Dertermine at what period of the year we should enter the US sports good market 

# We need firstly to dertime the NAICS US code for our activity 

path1 = './NAICS/2017 NAICS Definition.csv'
NAICS_def = pd.read_csv(path1)
df_NAICS_def = pd.DataFrame(NAICS_def)
df_NAICS_def.info()  

df_NAICS_def.loc[df_NAICS_def['2017 NAICS US Code'] == '45111']

# our NAICS code is 45111
```

```python
# Using this code we will gather information for the past three years to 
# identify cycles 

# first we need to clean the data 

path = './NAICS/mrtssales92-present.xls'
NAICS = pd.read_excel(path, sheet_name=['2017','2018','2019'])

# get the 3 years 
NAICS_2017 = NAICS.get('2017')
NAICS_2018 = NAICS.get('2018')
NAICS_2019 = NAICS.get('2019')

# create data frame fort the three years 
df_NAICS_2017 = pd.DataFrame(NAICS_2017)
df_NAICS_2018 = pd.DataFrame(NAICS_2018)
df_NAICS_2019 = pd.DataFrame(NAICS_2019)

# check if datatype is constant -> it is the case 
df_NAICS_2017.info() # identify that NAICS code has a double space 
df_NAICS_2018.info()# identify that NAICS code has a double space
df_NAICS_2019.info()# identify that NAICS code has a double space

#reference correctly total colunms 
#df_NAICS_2017["Total_2017"] = df_NAICS_2017["TOTAL"]
#df_NAICS_2018["Total_2018"] = df_NAICS_2018["TOTAL"]
#df_NAICS_2019["Total_2019"] = df_NAICS_2019["TOTAL"]

# drop unreferenced total columns 
df_NAICS_2017 =df_NAICS_2017.drop(columns=['TOTAL'])
df_NAICS_2018 =df_NAICS_2018.drop(columns=['TOTAL'])
df_NAICS_2019 =df_NAICS_2019.drop(columns=['TOTAL'])

# append three data frame to create one 
df_N_17_19 = pd.concat([df_NAICS_2017, df_NAICS_2018, df_NAICS_2019])
df_N_17_19


# Rename NAICS column correctly -> double space on the old name 
df_N_17_19["NAICS_c"] = df_N_17_19["NAICS  Code"]
df_N_17_19
df_NAICS_454111 =df_N_17_19.loc[df_N_17_19["NAICS_c"] == 45111]

#drop NAICS code useless and create the data frame that we will use for the analysis 
df_N_anl =df_NAICS_454111.drop(columns=['NAICS  Code'])
df_N_anl
```

```python
# Final table to analyze 

#condensing the data 

NAICS_melt = df_N_anl.melt(
    id_vars=['Kind of Business', 'NAICS_c'],
    var_name='month',
    value_name='TOTAL')
NAICS_melt

#drop NAN 
df_NAICS_dropped = pd.DataFrame.dropna(NAICS_melt)
df_NAICS_dropped
```

```python
# create the graph summurazing our data 

df_NAICS_dropped['month'] = pd.to_datetime(df_NAICS_dropped['month'])

import seaborn as sns
sns.set_theme(style="darkgrid")

ax=sns.lineplot(x="month", y="TOTAL",
             hue="Kind of Business", 
             data=df_NAICS_dropped)

ax.tick_params(axis='x', rotation=50)


# Second recommendation 

print(f"""
ASSUMPTION: the TOTAL column shows the number of stores opening for each month. 
Using the three years 2017, 2018, 2019 we determine a cycle.

Every extrem points on the graph correspond to the months ex: 1st = January.

We see that the minimum opening of stores correspond to 1st and 9th 10th month 
for every year. 
Meaning that January, September and October represent the down peak of stores opening. 
Where the highest peak is identified at the 12th month (December). 

Why are more stores  opening in december ? 
The advantage to start in december will be to start at full productivity approaching 
the festive period and represents in this sense an opportunity. 
We should hence avoid to lauching our new stores in january, september and october 
where the customer's willingness to buy is low. 

Answer the question when to enter the US market of sporting good stores. 


CONCLUSION: December is priviligied and avoid months where customer's 
willingness to buy is low such as January, September and October. 


""")
```

```python

rivalry = pd.read_csv('./noon.csv')
df_r = pd.DataFrame(rivalry)
df_r
```

```python
df_rcp = df_r.copy()
df_rcp = df_rcp.drop(columns= ['name','sold_by', 'current_price', 'saving', 'section', 'color', 'material'])
df_rcp = df_rcp.rename(columns= {'old_price' : 'price'})
df_rcp
```

```python
# selecting only our 4 sports brand amoung all the brands (after identifying every brands)
#only the brands focus on sports goods have been identifyied as direct competitors 
#ex: Skechers, le coq sportif etc.. are not mainly sport brand

df_rcp1 = df_rcp.loc[(df_rcp['brand'] == 'Nike')]
df_rcp2 = df_rcp.loc[(df_rcp['brand'] == 'adidas')]
df_rcp3 = df_rcp.loc[(df_rcp['brand'] == 'UNDER ARMOUR')]
df_rcp4 = df_rcp.loc[(df_rcp['brand'] == 'PUMA')]

df_rcp0 = pd.concat([df_rcp1, df_rcp2, df_rcp3, df_rcp4])

#identify the outliers 
plot = sns.histplot(data=df_rcp0,
                   x='price',
                    hue='brand')

# we found two outliers close to 600 
# we also see that NIKE and adidas are the two brands with the biggest quantity of products 
```

```python
# we are checking our the values of our outliers and ensure to a proper outliers selection
df_rcp0 = df_rcp0.sort_values(by = 'price', ascending = False)
df_rcp0.head(10)
# we detected 2 outliers row 83 and row 35  way too high and would false our average price 
```

```python
# now we will check the values of the tail to be sure that there is no outliers 
df_rcp0.tail(10)
# no outliers 
```

```python
# drop our two outliers 
df_rcp0 = df_rcp0.drop(
                    labels = [83 , 35],
                    axis = 0,
                    inplace = False 
                    )
```

```python
#ensure that the outliers have been dropped 
df_rcp0.head()
# it is the case 
```

```python
# now ce can calculate the average of every brand
# from this we will be able to establish the price strategy of our direct competitors
```

```python
# mean 
df_rcpmean = df_rcp0.groupby('brand').mean().sort_values(by = 'price', ascending = False)
df_rcpmean
```

```python
# median 
df_rcpmed = df_rcp0.groupby('brand').median().sort_values(by = 'price', ascending = False)
df_rcpmed
```

```python
print(f"""

ASSUMPTION: 
We are a sport company only producing goods for sport purposes.
For instance le coq sportif is identyfied by some as a sport goods brand, 
however this company is not mainly focused on producing good for sport purposes. 
Where, Puma, Nike Under Armour and Adidas are mainly focused on producing goods 
for sport purposes.
These four brands are in this sense 'Direct competitors'.

CONCLUSION:
We identified our direct competitors: Puma, Nike, Under Armour, Adidas. 
Crossing the information collected from the median and average (after eliminating outliers)
we are able to determine the price strategy of our direct competitors. 

Hence we encourage to follow the same price strategy for our company's expansion in the US. 

""")
```
